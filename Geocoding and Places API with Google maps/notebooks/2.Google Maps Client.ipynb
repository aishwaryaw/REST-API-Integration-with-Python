```python
import requests
import pandas as pd
from urllib.parse import urlencode
```

```python
api_key = "api-key"
```

```python
class GoogleMapsAPI():
    lat = None
    lng = None
    location_query = None
    api_key = None
    data_type = 'json'
    def __init__(self, api_key, address_or_location=None, *args, **kwargs):
        if api_key == None:
            raise Exception("u don't have api key")
        self.api_key = api_key
        self.location_query = address_or_location
        if self.location_query != None:
            self.extract_lat_and_lng()
    
    
    def extract_lat_and_lng(self, location=None):
        loc_query = self.location_query
        if location != None:
            loc_query = location
        params = {
            "address": loc_query,
            "key" : self.api_key
        }
        endpoint = f"https://maps.googleapis.com/maps/api/geocode/{self.data_type}"
        params_encode = urlencode(params)
        final_endpoint = f"{endpoint}?{params_encode}"
        r = requests.get(final_endpoint)
        data = r.json()
        if r.status_code not in (200,299):
            return {}
        latlng = {}
        try:
            latlng = data['results'][0]['geometry']['location']
        except:
            pass
        lat,lng = latlng.get('lat'), latlng.get('lng')
        self.lat = lat
        self.lng = lng

        return lat,lng
    
    
    def search_nearby_places(self, radius=5000 , keyword= "Mexican food", location=None):
        endoint = f"https://maps.googleapis.com/maps/api/place/nearbysearch/{self.data_type}"
        lat , lng = self.lat , self.lng
        if location != None:
            lat , lng = self.extract_lat_and_lng(location)
        params = {
            "key": self.api_key,
            "location": f"{lat}{lng}",
            "radius": radius,
            "keyword": keyword
        }
        params_encode = urlencode(params)
        final_places_endpoint = f"{endpoint}?{params_encode}"
        r = requests.get(final_endpoint)
        if r.status_code not in (200,299):
            return {}
        return r.json()
    
    
    def search_places(self, location= None , inputtype="textquery", _input="Mexican Food", fields=["place_id","formatted_address , geometry"],radius=5000, use_cicular = False):
        endpoint = f"https://maps.googleapis.com/maps/api/place/findplacefromtext/{self.data_type}"
        params = {
            "key": self.api_key,
            "input": _input,
            "inputtype": inputtype,
            "fields": fields
        }
        
        lat , lng = self.lat, self.lng
        if location != None:
            lat, lng = self.extract_lat_and_lng(location)

        locationbias = f"point:{lat}{lng}"
        if use_circular:
            locationbias = f"circle:{radius}@{lat}{lng}"

        params['locationbias'] = locationbias

        params_encode = urlencode(params)
        final_places_endpoint = f"{endpoint}?{params_encode}"
        r = requests.get(final_endpoint)
        if r.status_code not in (200,299):
            return {}
        return r.json()


    def place_detail(self, place_id, fields = ["name","rating","formatted_phone_number","formatted_address", "photo"]):

        endpoint = f"https://maps.googleapis.com/maps/api/place/details/{self.data_type}"
        params = {
            "key": self.api_key,
            "place_id": place_id,
            "fields": ",".join(fields)
        }

        params_encode = urlencode(params)
        final_places_endpoint = f"{endpoint}?{params_encode}"
        r = requests.get(final_endpoint)
        if r.status_code not in (200,299):
            return {}
        return r.json()
    
    
    def get_place_photo(self, photo_ref, max_height=400 , max_width=400 ):
#     https://maps.googleapis.com/maps/api/place/photo?maxwidth=400&photoreference=CnRtAAAATLZNl354RwP_9UKbQ_5Psy40texXePv4oAlgP4qNEkdIrkyse7rPXYGd9D_Uj1rVsQdWT4oRz4QrYAJNpFX7rzqqMlZw2h2E2y5IKMUZ7ouD_SlcHxYq1yL4KbKUv3qtWgTK0A6QbGh87GB3sscrHRIQiG2RrmU_jF4tENr9wGS_YxoUSSDrYjWmrNfeEHSGSc3FyhNLlBU&key=YOUR_API_KEY

        endpoint = "https://maps.googleapis.com/maps/api/place/photo"
        params = {
            "key": self.api_key,
            "photoreference": photo_ref,
            "maxheight":max_height,
            "maxwidth": max_width
        }

        params_encode = urlencode(params)

        final_endpoint = f"{endpoint}?{params_encode}"
        r = requests.get(final_endpoint)
        if r.status_code not in (200,299):
            raise Exception("Error !!")
        return r.json()
```
